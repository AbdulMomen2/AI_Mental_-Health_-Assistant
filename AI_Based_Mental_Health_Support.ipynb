```python
# Load models
whisper_model = whisper.load_model("medium", device=DEVICE)
tokenizer = AutoTokenizer.from_pretrained("saha6754/mental_health_finetuned")
model = AutoModelForCausalLM.from_pretrained("saha6754/mental_health_finetuned")
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python

def transcribe_audio(audio_path):
    audio = whisper.load_audio(audio_path)
    audio = whisper.pad_or_trim(audio)
    mel = whisper.log_mel_spectrogram(audio).to(DEVICE)
    _, probs = whisper_model.detect_language(mel)
    options = whisper.DecodingOptions()
    result = whisper.decode(whisper_model, mel, options)
    return result.text
```

```python
def generate_response(user_input):
    prompt = f"""
You are a supportive and compassionate mental health assistant.
Listen to the user's concern and respond with empathy and helpful guidance.

USER: {user_input}
ASSISTANT:
"""
    result = pipe(prompt, max_new_tokens=200, do_sample=True)
    return result[0]["generated_text"].split("ASSISTANT:")[-1].strip()
```

```python
def analyze_face(image_path):
    result = DeepFace.analyze(img_path=image_path, actions=['emotion'], enforce_detection=False)[0]
    emotion = result['dominant_emotion']
    confidence = result['emotion'][emotion]
    eye_info = "Eye movement can't be extracted directly but blinking can be inferred using OpenCV (advanced)"
    return emotion, confidence, eye_info

def generate_ecg_graph(emotion):
    # Simulate ECG based on emotion
    t = np.linspace(0, 1, 1000)
    ecg = np.sin(2 * np.pi * 5 * t)  # baseline ECG

    if emotion in ['sad', 'angry', 'fear']:
        ecg *= 0.5
    elif emotion == 'happy':
        ecg *= 1.5

    fig, ax = plt.subplots()
    ax.plot(t, ecg)
    ax.set_title(f"Simulated ECG based on emotion: {emotion}")
    ax.set_xlabel("Time")
    ax.set_ylabel("Amplitude")

    temp_plot_path = tempfile.mktemp(suffix=".png")
    fig.savefig(temp_plot_path)
    plt.close(fig)
    return temp_plot_path
```

```python
def mental_health_chatbot(audio, text_input, image):
    user_text = ""

    if audio is not None:
        with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as temp_audio:
            temp_audio.write(audio.read())
            user_text = transcribe_audio(temp_audio.name)
    elif text_input:
        user_text = text_input
    else:
        user_text = "No input provided"

    response = generate_response(user_text)
    audio_response = text_to_speech(response)

    emotion_result = "No image provided"
    ecg_plot = None

    if image is not None:
        temp_img_path = tempfile.mktemp(suffix=".jpg")
        image.save(temp_img_path)
        emotion, confidence, eye_info = analyze_face(temp_img_path)
        emotion_result = f"Emotion: {emotion}, Confidence: {confidence:.2f}%, Eye Status: {eye_info}"
        ecg_plot = generate_ecg_graph(emotion)

    return response, audio_response, emotion_result, ecg_plot
```

```python
import torch
import whisper
from transformers import pipeline
from gtts import gTTS
import matplotlib.pyplot as plt
import numpy as np
import tempfile
import gradio as gr
import cv2
from PIL import Image

# Load models
whisper_model = whisper.load_model("base")
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
emotion_classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", top_k=5)

# Transcribe audio
def transcribe_audio(audio_path):
    try:
        audio = whisper.load_audio(audio_path)
        audio = whisper.pad_or_trim(audio)
        mel = whisper.log_mel_spectrogram(audio).to(DEVICE)
        options = whisper.DecodingOptions()
        result = whisper.decode(whisper_model, mel, options)
        return result.text
    except Exception as e:
        return f"Transcription error: {e}"

# Chatbot response
def generate_response(user_input):
    try:
        prompt = f"You are a compassionate mental health assistant.\nUser: {user_input}\nAssistant:"
        result = pipe(prompt, max_new_tokens=150, do_sample=True)
        return result[0]["generated_text"].split("Assistant:")[-1].strip()
    except Exception as e:
        return f"Response error: {e}"

# Text-to-speech
def text_to_speech(text, filename="response.mp3"):
    try:
        tts = gTTS(text)
        tts.save(filename)
        return filename
    except Exception as e:
        print(f"TTS error: {e}")
        return None

# Text emotion detection
def analyze_text_emotion(text):
    try:
        results = emotion_classifier(text)[0]
        top_emotion = results[0]['label']
        top_conf = results[0]['score'] * 100
        summary = "\n".join([f"{res['label']}: {res['score']*100:.2f}%" for res in results])
        return top_emotion, f"Text Emotion: {top_emotion} (Confidence: {top_conf:.2f}%)\n{summary}"
    except Exception as e:
        return "neutral", f"Text analysis error: {e}"

# ECG graph based on emotion
def generate_ecg_graph(emotion):
    try:
        t = np.linspace(0, 1, 1000)
        ecg = np.sin(2 * np.pi * 5 * t)
        emotion_multipliers = {
            "sad": 0.5, "angry": 0.6, "fear": 0.5,
            "happy": 1.5, "surprise": 1.3,
            "neutral": 1.0, "disgust": 0.4
        }
        ecg *= emotion_multipliers.get(emotion.lower(), 1.0)

        fig, ax = plt.subplots()
        ax.plot(t, ecg, linestyle='--', marker='o', markersize=2)
        ax.set_title(f"🫀 Simulated ECG Based on Emotion: {emotion}")
        ax.set_xlabel("Time")
        ax.set_ylabel("Amplitude (Emotion Signal)")
        ax.grid(True)

        path = tempfile.mktemp(suffix=".png")
        fig.savefig(path)
        plt.close(fig)
        return path
    except Exception as e:
        print("Graph error:", e)
        return None

# Analyze video emotions (frame-by-frame)
def analyze_video_emotion(video_path):
    emotions = []
    try:
        cap = cv2.VideoCapture(video_path)
        frame_skip = 10
        count = 0
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            if count % frame_skip == 0:
                temp_img_path = tempfile.mktemp(suffix=".jpg")
                cv2.imwrite(temp_img_path, frame)
                result = emotion_classifier(Image.open(temp_img_path).tobytes().decode(errors='ignore'))[0]
                emotions.append(result[0]['label'])
            count += 1
        cap.release()
        # Simple majority vote or top emotion count
        if emotions:
            from collections import Counter
            most_common = Counter(emotions).most_common(1)[0]
            return most_common[0], f"Video Emotion: {most_common[0]} (Most frequent across frames)"
        return "neutral", "No emotion detected in video"
    except Exception as e:
        print("Video analysis error:", e)
        return "neutral", f"Video analysis error: {e}"

# Main chatbot logic
def mental_health_chatbot(audio, text_input, video):
    try:
        user_text = ""
        if audio is not None:
            with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as temp_audio:
                temp_audio.write(audio.read())
                user_text = transcribe_audio(temp_audio.name)
        elif text_input:
            user_text = text_input
        else:
            user_text = "No input provided."

        # Generate response
        response = generate_response(user_text)
        audio_response = text_to_speech(response)

        # Text emotion
        text_emotion, text_emotion_summary = analyze_text_emotion(user_text)
        ecg_plot = generate_ecg_graph(text_emotion)

        # Video emotion
        video_emotion_summary = "No video provided"
        if video is not None:
            temp_video_path = tempfile.mktemp(suffix=".mp4")
            with open(temp_video_path, "wb") as f:
                f.write(video.read())
            _, video_emotion_summary = analyze_video_emotion(temp_video_path)

        full_emotion_report = f"{text_emotion_summary}\n\n{video_emotion_summary}"

        return response, audio_response, full_emotion_report, ecg_plot
    except Exception as e:
        return f"Flow error: {e}", None, "Error", None

# Gradio interface
iface = gr.Interface(
    fn=mental_health_chatbot,
    inputs=[
        gr.Audio(type="filepath", label="Speak or Upload Audio"),
        gr.Textbox(placeholder="Type your message", label="Text Input"),
        gr.Video(label="Upload Video (for Emotion Detection)"),
    ],
    outputs=[
        gr.Textbox(label="Chatbot Response"),
        gr.Audio(label="Voice Response"),
        gr.Textbox(label="Emotion Detection + Confidence"),
        gr.Image(label="Simulated ECG Graph"),
    ],
    title="🧠 Mental Health Chatbot: Text, Voice & Video Emotion + ECG",
    description="Talk or write to a mental health bot. Upload a video to analyze emotional tone across frames. Get ECG-style visualization and confidence-level emotion insights.",
    live=False
)

iface.launch()

```
